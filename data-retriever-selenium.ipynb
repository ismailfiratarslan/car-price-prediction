```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementClickInterceptedException, NoSuchElementException
import time
import pandas as pd
```

```python
path = "C:/Users/firatarslan/chromedriver.exe"
browser = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
browser.get("https://www.arabam.com/ikinci-el/otomobil?city=85&city=86&take=50")


def dataRetriever(index):

    time.sleep(3)

    allCars = browser.find_elements(By.TAG_NAME, 'tr')
    ids = []

    for id in allCars:
        ids.append(id.get_attribute("id"))
    while ('' in ids):
        ids.remove('')

    clickedElement = browser.find_element(By.ID, ids[index])
    clickedElement.click()

    specs = browser.find_elements(By.CSS_SELECTOR, '.bcd-list-item')
    prices = browser.find_elements(By.XPATH, '/html/body/div[2]/div[2]/div[3]/div/div[1]/div[1]/div[2]/div[1]/div/span')

    details = []
    price = []

    for i in prices:
        price.append(i.text)
        
    for i in specs:
        details.append(i.text)

    infos = []
    for x in details:
        infos.append(x.replace("\n", ""))
        
    infos = [item.split(':') for item in infos]
    

    strippedPrice = price[0].strip()
    strippedPrice = strippedPrice.replace(' TL', '')

    i = 0
    rows = []
    column = []

    while i <= len(infos)-1:
        rows.append(infos[i][1])
        column.append(infos[i][0])
        i += 1
    column.append("TL")

    strippedPrice = price[0].strip()
    strippedPrice = strippedPrice.replace(' TL', '')
    rows.append(strippedPrice)

    dfContents = pd.DataFrame(rows).T
    dfContents.columns = column

    dfContents.to_csv(r"cars.csv",encoding="utf-8-sig",index=False, mode="a")
    browser.execute_script("window.history.go(-1)")

def cars():
    index = 0
    
    while index<=50: 
        try: 
            dataRetriever(index)
            index += 1
        
        except IndexError:
            index += 1
            browser.execute_script("window.history.go(-1)")
            pass
        except ElementClickInterceptedException:
            index += 1
            browser.execute_script("window.history.go(-1)")
            pass
        except NoSuchElementException:
            index += 1
            browser.execute_script("window.history.go(-1)")
            pass

    if index == 51: 
        browser.execute_script("window.history.go(+1)")
        nextPage = browser.find_element(By.XPATH, '//*[@id="pagingNext"]')
        nextPage.click()
        cars()
        

cars()
    
        
```
